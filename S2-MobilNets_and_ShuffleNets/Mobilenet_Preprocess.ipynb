## **Rename all images to .jpg extension and remove files other than images**

```
import os
folders = ['Flying Birds','Small QuadCopters', 'Large QuadCopters', 'Winged Drones']
path = '/content/dataset/'

for folder in folders:
    data = path + folder
    for filename in os.listdir(data) :

      #if jpg continue
      if(filename.endswith('.jpg')):
        continue

      # else rename
      elif(filename.endswith('.png')):
        dst =filename.split('.png')[0] + '.jpg'

      elif(filename.endswith('.jpeg')):
        dst =filename.split('.jpeg')[0] + '.jpg'

      elif(filename.endswith('.webp')):
        dst =filename.split('.webp')[0] + '.jpg'

      elif(filename.endswith('.JPG')):
        dst =filename.split('.JPG')[0] + '.jpg'

      elif(filename.endswith('.JPEG')):
        dst =filename.split('.JPEG')[0] + '.jpg'

      elif(filename.endswith('.PNG')):
        dst =filename.split('.PNG')[0] + '.jpg'


    #if it is other than images - remove the file
      else:
        os.remove(data + filename)


      src =data + filename
      dst =data + dst
      os.rename(src, dst)
```

## **Rename all files with classname_<int_number>.jpg**

```
import os
classes =  ['Flying Birds','Small QuadCopters', 'Large QuadCopters', 'Winged Drones']
path = '/content/dataset/'
for classs in classes:

  data = path + classs
  i = 1
  f_name = 'Flying_Birds'
  for filename in os.listdir(data) :
    dst = f_name +"_"+ str(i) + '.jpg'
    i+=1
    src = data + filename
    dst = data + dst
    os.rename(src, dst)
```

## **Compute Mean and Standard Deviation of dataset**

```
#calculate STd and Mean
from tqdm import notebook
from PIL import Image
import numpy as np
def mean_std():

    n = 0
    s = np.zeros(3)
    sq = np.zeros(3)
    folders = ['Flying_Birds', 'Large_QuadCopters', 'Small_QuadCopters', 'Winged_Drones']
    path = '/content/dataset/'

    for folder in folders:
      print(folder)
      dir = path + folder

      for filename in notebook.tqdm(os.listdir(dir)):

            # Since webp, png may give channel error while computing, better to convert to 'RGB \
            # the time of opening the image only          
            img = Image.open(dir + '/' + filename).convert('RGB')
            x = np.array(img)/255
            s += x.sum(axis=(0,1))
            sq += np.sum(np.square(x), axis=(0,1))
            n += x.shape[0]*x.shape[1]
     


    mu = s/n
    std = np.sqrt((sq/n - np.square(mu)))
    print(mu, sq/n, std, n)
```

```

import os
import zipfile

def zipdir(path, ziph):
    # ziph is zipfile handle
    for root, dirs, files in os.walk(path):
        for file in files:
            ziph.write(os.path.join(root, file))


zipf = zipfile.ZipFile('session_2.zip', 'w', zipfile.ZIP_DEFLATED)
zipdir('dataset/', zipf)
zipf.close()
```
